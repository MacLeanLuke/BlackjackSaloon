```python
### IMPORT ###
import random

### INITIALIZE ###

# list of cards - unsorted
card_list = ["A_Spades", "A_Hearts", "A_Clubs", "A_Diamonds",
            "2_Spades", "2_Hearts", "2_Clubs", "2_Diamonds",
            "3_Spades", "3_Hearts", "3_Clubs", "3_Diamonds",
            "4_Spades", "4_Hearts", "4_Clubs", "4_Diamonds",
            "5_Spades", "5_Hearts", "5_Clubs", "5_Diamonds",
            "6_Spades", "6_Hearts", "6_Clubs", "6_Diamonds",
            "7_Spades", "7_Hearts", "7_Clubs", "7_Diamonds",
            "8_Spades", "8_Hearts", "8_Clubs", "8_Diamonds",
            "9_Spades", "9_Hearts", "9_Clubs", "9_Diamonds",
            "10_Spades", "10_Hearts", "10_Clubs", "10_Diamonds",
            "J_Spades", "J_Hearts", "J_Clubs", "J_Diamonds",
            "Q_Spades", "Q_Hearts", "Q_Clubs", "Q_Diamonds",
            "K_Spades", "K_Hearts", "K_Clubs", "K_Diamonds"]

# randomize the cards
random.shuffle(card_list)

# assign values to the cards
values = {"A_Spades":1, "A_Hearts":1, "A_Clubs":1, "A_Diamonds":1,
            "2_Spades":2, "2_Hearts":2, "2_Clubs":2, "2_Diamonds":2,
            "3_Spades":3, "3_Hearts":3, "3_Clubs":3, "3_Diamonds":3,
            "4_Spades":4, "4_Hearts":4, "4_Clubs":4, "4_Diamonds":4,
            "5_Spades":5, "5_Hearts":5, "5_Clubs":5, "5_Diamonds":5,
            "6_Spades":6, "6_Hearts":6, "6_Clubs":6, "6_Diamonds":6,
            "7_Spades":7, "7_Hearts":7, "7_Clubs":7, "7_Diamonds":7,
            "8_Spades":8, "8_Hearts":8, "8_Clubs":8, "8_Diamonds":8,
            "9_Spades":9, "9_Hearts":9, "9_Clubs":9, "9_Diamonds":9,
            "10_Spades":10, "10_Hearts":10, "10_Clubs":10, "10_Diamonds":10,
            "J_Spades":10, "J_Hearts":10, "J_Clubs":10, "J_Diamonds":10,
            "Q_Spades":10, "Q_Hearts":10, "Q_Clubs":10, "Q_Diamonds":10,
            "K_Spades":10, "K_Hearts":10, "K_Clubs":10, "K_Diamonds":10}

# to check if the card is an Ace
aces = ["A_Spades", "A_Hearts", "A_Clubs", "A_Diamonds"]


# to check if the card is a 10, Jack, Queen or King
tens = ["10_Spades", "10_Hearts", "10_Clubs", "10_Diamonds",
            "J_Spades", "J_Hearts", "J_Clubs", "J_Diamonds",
            "Q_Spades", "Q_Hearts", "Q_Clubs", "Q_Diamonds",
            "K_Spades", "K_Hearts", "K_Clubs", "K_Diamonds"]

# initialize cards at player and dealer to be empty list
playercard = []
dealercard = []


# condition for the player's turn
toContinue = True


#flag to see if the player wins, loses, or draws
playerWin = "Unknown"



### DEFINE ###

# function to draw cards from deck of cards
def takecard(mycard):
    mycard.append(card_list[0])
    card_list.pop(0)
    

# function to check if the card is an Ace
def check_aces(val):
    for i in range(len(aces)):
        if val == aces[i]:
            return True
        
        
#function to check if the card is a 10, Jack, Queen or King
def check_tens(val):
    for i in range(len(tens)):
        if val == tens[i]:
            return True

        
# function to check if the two cards are Blackjack
def is_blackjack(mycards):
    if (check_aces(mycards[0]) and check_aces(mycards[1])):
        return "Blackace"
    elif (check_aces(mycards[0]) and check_tens(mycards[1])):
        return "Blackjack"
    elif (check_tens(mycards[0]) and check_aces(mycards[1])):
        return "Blackjack"
    else:
        return "normal"
      
        
# Function to allow Ace to take either value 1 or 11
def values_of_Aces(mycard, my_sum):
    for i in range(len(mycard)):
        if check_aces(mycard[i]):
            my_sum += 10
            if my_sum > 21:
                my_sum -= 10
    return my_sum


# Function to add sum of cards for each card on hand
def sum_cards(mycard):
    my_sum = 0
    for i in range(len(mycard)):
        my_sum += values[mycard[i]]
    my_sum = values_of_Aces(mycard, my_sum)
    if len(mycard)==2:
        if check_aces(mycard[0]) and check_aces(mycard[1]):
            my_sum = 21
    return my_sum



### PLAY ###

#Print welcome sentance
print("WELCOME TO THE BLACKJACK SALOON PARTNER")
#playerName = input("What's your name?")
input("Press <enter> and come on in")


#Print the shuffled cards list, comment out for real play
# print("\nOriginal shuffled cards:\n", card_list)


#Player takes first card, dealer second, player third, dealer fourth
takecard(playercard)
takecard(dealercard)
takecard(playercard)
takecard(dealercard)



### TESTING HERE ###
# playercard = ["A_Spades", "A_Hearts"]
# dealercard = ["A_Hearts", "K_Spades"]



print("\nYour cards:", playercard)
print("Your Cards Value:", sum_cards(playercard))
print("\nThe Dealer's cards:", "[x]", dealercard[1])
print("Dealer Card Value: unknown \n")


# to check if there is a blackjack
if is_blackjack(playercard) != "normal" and is_blackjack(dealercard) != "normal":
    print("\nIt's a draw Partner.")
    toContinue = False
    playerWin = "Draw"
elif is_blackjack(playercard) != "normal":
    print("\nYou win Partner! Blackjack!")
    toContinue = False
    playerWin = "Win"
elif is_blackjack(dealercard) != "normal":
    print("\nTry again, Partner. Dealer wins!")
    toContinue = False
    playerWin = "Lose"
    
    
# Player's turn
while toContinue:
    answer = input("\nYou wanna Hit (H) or Stand (S) Partner?")
    if answer == "H":
        takecard(playercard)
        print("\nYour cards:", playercard)
        print("Your card value: ", sum_cards(playercard))
        if sum_cards(playercard)>21:
            print("\nToo bad! You Bust! \nTry Again Partner! ")
            toContinue = False
            playerWin = "Lose"
            break
        elif sum_cards(playercard) == 21:
            print("\nLucky Duck! \nYou got Blackjack!")
            toContinue = False
            playerWin = "Win"
            break
        elif len(playercard) == 5:
            print("\nYou have 5 cards. \nYou got lucky this time! \nYou Win Partner!")
            toContinue = False
            playerWin = "Win"
            break
    elif answer == "S":
        print("\nAlrighty, you stand.")
        print("\nYou have:", playercard)
        print("Your Total is: ", sum_cards(playercard))
        toContinue = False
        break
        

# Dealer's Turn
if playerWin == "Unknown":
    while True:
        if sum_cards(dealercard) < sum_cards(playercard):
            takecard(dealercard)
            print("\nDealer has:", dealercard)
            print("The sum is:", sum_cards(dealercard))
        else:
            print("\nFinal cards for Dealer:", dealercard)
            print("Final sum for Dealer:", sum_cards(dealercard))
            break
    
    # to check who wins!
    if sum_cards(dealercard) > 21:
        print("\nThe Dealer busted! \nYou got lucky this time... \nYou Win!!")
        playerWin = "Win"
        
    elif sum_cards(dealercard) > sum_cards(playercard):
        print("\nYa know the house always wins! \nTry Again Partner!!")
        playerWin = "Lose"
        
    elif sum_cards(dealercard) < sum_cards(playercard):
        print("\nYou got an Ace up your Sleeve? \nYou win Partner!")
        playerWin = "Win"
    
    elif sum_cards(dealercard) == sum_cards(playercard):
        print("\nIt's a draw Partner!")
        playerWin = "Draw"

#print("\nRemaining cards:\n", card_list)
print("\nThere are ", len(card_list), "left.")
```
